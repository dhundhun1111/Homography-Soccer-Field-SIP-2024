```python
from skimage import io, color, feature, transform
import matplotlib.pyplot as plt
import numpy as np
import time

# Load the image
image_path = "./assets/world_cup_template.png"
image = io.imread(image_path)

# Convert the image to grayscale
gray_image = color.rgb2gray(image)

# Perform edge detection using Canny algorithm
edges = feature.canny(gray_image, sigma=2.0)

# Perform Hough Transform to detect lines
hough_result = transform.probabilistic_hough_line(edges, threshold=10, line_length=50, line_gap=10)

# Display the results
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].imshow(edges, cmap='gray')
ax[0].set_title('Canny Edges')
ax[0].set_axis_off()

ax[1].imshow(image)
for line in hough_result:
    p0, p1 = line
    ax[1].plot((p0[0], p1[0]), (p0[1], p1[1]), '-r')
ax[1].set_title('Probabilistic Hough Transform')
ax[1].set_axis_off()

plt.show()

# Print the points detected
hough_result
```

```python
# Function to calculate the midpoint of a line segment
def midpoint(p0, p1):
    return ((p0[0] + p1[0]) / 2, (p0[1] + p1[1]) / 2)

# Calculate midpoints for all detected lines
midpoints = [midpoint(p0, p1) for p0, p1 in hough_result]

midpoints
```

```python
# take random sets of 4 corresponding points from the midpoints and midpoints2
import random
import numpy as np

# Randomly Select Different combinations of 4 corresponding points
def get_random_points(midpoints, midpoints2):
    random.seed(time.time())
    indices = random.sample(range(len(midpoints)), 4)
    points1 = np.array([midpoints[i] for i in indices])
    random.seed(time.time())
    indices = random.sample(range(len(midpoints)), 4)
    points2 = np.array([midpoints2[i] for i in indices])
    return

get_random_points(midpoints, midpoints2)

# Display the selected points
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].imshow(image)
ax[0].plot(points1[:, 0], points1[:, 1], 'ro')
ax[0].set_title('Selected Points in Image 1')
ax[0].set_axis_off()

ax[1].imshow(image2)
ax[1].plot(points2[:, 0], points2[:, 1], 'ro')
ax[1].set_title('Selected Points in Image 2')
ax[1].set_axis_off()

plt.show()

```

```python
# Function to calculate the homography matrix
def calculate_homography_matrix(points1, points2):
    A = []
    for i in range(4):
        x1, y1 = points1[i]
        x2, y2 = points2[i]
        A.append([-x1, -y1, -1, 0, 0, 0, x1*x2, y1*x2, x2])
        A.append([0, 0, 0, -x1, -y1, -1, x1*y2, y1*y2, y2])
    A = np.array(A)
    U, S, V = np.linalg.svd(A)
    H = V[-1].reshape(3, 3)
    return H

# Calculate the homography matrix
H = calculate_homography_matrix(points1, points2)

H

# Function to apply homography transformation
def apply_homography_transform(image, H):
    rows, cols = image.shape[:2]
    new_image = np.zeros_like(image)
    for x in range(cols):
        for y in range(rows):
            p = np.array([x, y, 1])
            q = np.dot(H, p)
            q = (q / q[2])[:2]
            if 0 <= q[0] < cols and 0 <= q[1] < rows:
                new_image[y, x] = image[int(q[1]), int(q[0])]
    return new_image

# Apply homography transformation to the image
warped_image = apply_homography_transform(image, H)

# Display the results
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].set_axis_off()

ax[1].imshow(warped_image)
ax[1].set_title('Warped Image')
ax[1].set_axis_off()

plt.show()
```

```python
# find the error value between the images
def calculate_error(image1, image2):
    return np.sum((image1 - image2) ** 2)

# Calculate the error value
error = calculate_error(image, warped_image)

error
```
